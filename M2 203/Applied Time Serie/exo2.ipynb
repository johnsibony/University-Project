```python
# Computing Exercise 2 : Absorption Ratio

# Below all the necessary packages (based on my correction)
from pandas import *
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# load data from "Data4Exam_Ex3.csv" into Data4Exam_Ex3 (type DataFrame)
Data4Exam_Ex3 = pandas.read_excel('Data4Assignment_CompEx2.xlsx', sheetname=None)

# In dataSP500, retrieve the data from the worksheet 'S&P500 cours', parse instruction
dataSP500 = Data4Exam_Ex3["S&P500 cours"]

# In dataP, retrieve the data from the worksheet 'industries_cours', parse instruction
dataP = Data4Exam_Ex3["industries_cours"]

# Create T and N containing respectively the number of observations (rows) and series (columns) of dataP
T = len(dataP)
N = len(dataP.columns)

# Create datesP containing only the variable "date" of dataP
datesP = dataP["date"]

# Remove the variable "date" from dataP (keep only the index and stock prices)
dataP = dataP.drop("date", axis=1)

# Create dataR from dataP containing the logarithmic returns of stocks
dataR = np.log(dataP)-np.log(dataP.shift(1))

# Do not keep the first row of dataR in order to withdraw NaN
dataR = dataR[1:]

# Create Tr and Nr respectively containing the number of observations (rows) and series (columns) of dataR
Tr = len(dataR)
Nr = len(dataR.columns)

# Create a scalar window containing the size of the rolling window
window = 500

# Create period containing a vector column from from 1 to window 
period = np.arange(1, window+1)

# Create vectorW containing a vector column of dimension (window,1) filled with window
vectorW = window * np.ones(window)

# Create weights containing a vector column of dimension (window,1) filled with exp( (-vecteurW-period)*log(2)/(vectorW/2) ), see Equation 5
weights = np.exp( (- (vectorW - period) * np.log(2) / (vectorW / 2) ))

# Shrink the values of weights by dividing them by its sum
weights = weights / weights.sum()

# Duplicate this last vector column weights to get a dataframe of dimension (window,Nr)
weights = pandas.DataFrame(weights, index=np.arange(1,window+1))
for i in range(Nr-1):
    weights[str(i+1)] = weights[0]

# Create a normalize function to standardize your data
def my_normalize(data):
    
    mu = data.mean()
    sigma = data.std()
    mylist = (data - mu) / sigma
    
    return data

# Initialize the dataframe AR of dimension (Tr-window) containing only NaN and two vector columns 'Date' and 'AR'
AR = pandas.DataFrame(np.nan, index=np.arange(Tr-window), columns=['Date','AR'], dtype='float')

#Loop from 1 to Tr-window (or from 0 to Tr-window-1)
for i in range(Tr-window):

    # Create a dataframe temp of dimension (windows,Nr) selecting only the observations of the dataframe dataR*weights, which will be used to run the ACP
    temp = dataR[i:i+window].multiply(weights.squeeze().values)
    
    # PCA instruction allowing to create pca in which we ask the computation of Nr components
    pca = PCA(n_components=Nr)

    # Report in results_pca the results of PCA on standardized data temp
    results_pca = pca.fit(my_normalize(temp))

    # Report the i-th observation of the 1+windows+i date in column 'Date'
    AR['Date'][i] = datesP.iloc[i+window+1]

    # Report the variability fraction explained by the first Nr/5 principal components, see Equation 4
    AR['AR'][i] = np.sum(results_pca.explained_variance_ratio_[0:int(Nr/5)])
    print(AR['AR'][i])

# Plot Figure 1
fig, ax1 = plt.subplots()
ax1.plot(AR['Date'], AR['AR'], 'b', label="Absorption Ratio")
ax1.set_ylabel('Absorption Ratio')
ax1.tick_params('y', colors='b')
ax1.set_ylim([0.6,2])
plt.legend(loc='upper right', bbox_to_anchor=(0.5, -.15))

ax2 = ax1.twinx()
ax2.plot(dataSP500["Date"], dataSP500["Close"], 'r', label="S&P500")
ax2.set_ylabel('Price')
ax2.set_ylim([0,1600])
plt.legend(loc='upper left', bbox_to_anchor=(0.55, -0.15))

ax1.set_xlabel('Time')
ax1.set_xlim('1998', '2010')

plt.show()
```
